```python
from gi.repository import Gtk

class window:

    def __init__(self, operations):

        self.builder = Gtk.Builder()
        self.builder.add_from_file('stat_win.glade')

        self.count_total = self.builder.get_object('button2')
        self.count_r = self.builder.get_object('button21')
        self.count_i = self.builder.get_object('button22')
        self.count_j = self.builder.get_object('button23')
        self.count_ls = self.builder.get_object('button24')
        self.count_o = self.builder.get_object('button25')
        self.percent_r = self.builder.get_object('button31')
        self.percent_i = self.builder.get_object('button32')
        self.percent_j = self.builder.get_object('button33')
        self.percent_ls = self.builder.get_object('button34')
        self.percent_o = self.builder.get_object('button35')

		self.r_type = ["add", "sub", "and", "or", "xor", "nor", "slt", "sltu"]
		self.i_type = ["addi", "subi", "slti", "sltiu","andi", "ori", "xori", "lui", "beq", "sll", "srl", "sra"]
		self.j_type = ["j", "jal", "jr"]
		self.ls_type = ["lw", "sw"]
		self.others = ["print", "exit", "nop"]

        self.total = 0
        self.r = 0
        self.i = 0
        self.j = 0
        self.ls = 0
        self.o = 0

        for op in operations:
            if op is not None:
                self.total += 1
                if op in r_type:
                    self.r += 1
                elif op in i_type:
                    self.i += 1
                elif op in j_type:
                    self.j += 1
                elif op in ls_type:
                    self.ls += 1
                else:
                    self.o += 1

        self.window = self.builder.get_object('window1')
        self.window.set_title('Instruction Statistics')

        self.window.connect('delete-event', Gtk.main_quit)
        self.window.show_all()
        Gtk.main()
        
```
